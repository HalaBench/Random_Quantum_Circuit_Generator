```python
import random
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.circuit.library import XGate, YGate, ZGate, HGate, CXGate,CYGate,IGate, CZGate
```

```python
def random_circuit(num_qubits, depth, basis_gates):
    
    #create a quantum circuit with a certain number of qubits
    qr = QuantumRegister(num_qubits)
    qc = QuantumCircuit(qr)
    
    #qbit indexes
    length = num_qubits-1
    
    
    #if there are no qubits in the circuit we return an empty circuit
    if(num_qubits == 0):
        return qc       
    
    else:
        
        #while we didn't reach the wanted depth we can keep going
        while (qc.depth() != depth):
        
            #get a gate randomely from the provided list of gates
            gate_name = random.choice(basis_gates)
        
        
            #if we have a single qubit we can't work with 2 qubit gates
            if(num_qubits == 1):
                                        
                if (gate_name == "x"):
                    qc.append(XGate(), [0])              
                elif (gate_name == "y"):
                    qc.append(YGate(), [0])   
                elif (gate_name == "z"):
                    qc.append(ZGate(), [0])   
                elif (gate_name == "h"):
                    qc.append(HGate(), [0])
                else:
                    qc.append(IGate(), [0])
            else:
                
                    
                    qIndex, Target_qIndex = random.sample(list(range(0, num_qubits)), 2)
                              
                    if (gate_name == "x"):
                        qc.append(XGate(), [qIndex])              
                    elif (gate_name == "y"):
                        qc.append(YGate(), [qIndex])   
                    elif (gate_name == "z"):
                        qc.append(ZGate(), [qIndex])   
                    elif (gate_name == "h"):
                        qc.append(HGate(), [qIndex]) 
                    elif (gate_name == "cx"):           
                        qc.append(CXGate(), [qIndex, Target_qIndex])                 
                    elif (gate_name == "cy"):             
                        qc.append(CYGate(), [qIndex, Target_qIndex])
                    elif (gate_name == "cz"):             
                        qc.append(CZGate(), [qIndex, Target_qIndex])
                    else:
                        qc.append(IGate(), [qIndex])
    
    
    return qc
```

# Test the code

```python
qc = random_circuit(2,10,['x', 'y', 'z', 'h','cx','cz'])
qc.draw(output='mpl')
```

```python
print("Circuit depth: ", qc.depth())
```
